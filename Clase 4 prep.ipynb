```python
rng = pd.date_range(pd.Timestamp("2018-03-10 09:00"),periods=3, freq='s')
rng.strftime('%B %d, %Y, %r')

# pd.date_range es una función de pandas que permite generar un rango de fechas con una frecuencia específica. 
# Los argumentos que se le pasan son el punto de inicio y el número de períodos (en este caso, 3) y la frecuencia 
# (en este caso 's' para segundos).

# La función pd.Timestamp convierte una fecha y hora en un objeto de fecha y hora de pandas. 
# En este caso, se está creando un Timestamp con la fecha "2018-03-10 09:00".

# Entonces, pd.date_range(pd.Timestamp("2018-03-10 09:00"),periods=3, freq='s') 
# genera un rango de fechas con un período de 3 segundos a partir de "2018-03-10 09:00".

# Finalmente, el método strftime se está utilizando para dar formato a las fechas en el rango,
#  donde '%B %d, %Y, %r' es un formato personalizado para mostrar el mes en texto, el día, el año y
#  la hora en formato de 12 horas con minutos y segundos.
```
